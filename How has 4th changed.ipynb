```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df2017 = pd.read_csv('plays_4thdowns.csv')
df2019 = pd.read_csv('reg_pbp_2019.csv')
dfSim = pd.read_csv('generated_data.csv')
```

```python
plt.hist(df2017['yardlineNumber'], bins=20, alpha=0.5, label='2017')
plt.hist(df2019['yardline_half'], bins=20, alpha=0.5, label='2019')
#plt.hist(dfSim['yardlineNumber'], bins=20, alpha=0.5, label='sim')

plt.legend(loc='upper right')
#plt.save('location_on_field_hist.jpg')
```

```python
#the density represents the estimated probability density of the data at each point along the x-axis.
#It can be interpreted as the relative likelihood of observing a data point at that value.
sns.kdeplot(df2017['yardlineNumber'], label='2017')
sns.kdeplot(df2019['yardline_half'], label='2019')
#sns.kdeplot(dfSim['yardlineNumber'], label='sim')

plt.legend(loc='upper right')
#plt.save('location_on_field_kde.jpg')
```

```python
df2017_4th_down_success = df2017[(df2017['down'] == 4) & (df2017['firstDown_Conversion'] == 1)]
df2019_4th_down_success = df2019[(df2019['down'] == 4) & (df2019['fourth_down_converted'] == 1)]
dfSim_4th_down_success = dfSim[(dfSim['down'] == 4) & (dfSim['firstDown_Conversion'] == 1)]


df2017_conversion_rate = len(df2017_4th_down_success) / len(df2017[df2017['down'] == 4])
df2019_conversion_rate = len(df2019_4th_down_success) / len(df2019[df2019['down'] == 4])
dfSim_conversion_rate = len(dfSim_4th_down_success) / len(dfSim[dfSim['down'] == 4])

print('Conversion rate for 4th down in DataFrame 1:', df2017_conversion_rate)
print('Conversion rate for 4th down in DataFrame 2:', df2019_conversion_rate)
print('Conversion rate for 4th down in DataFrame 1:', dfSim_conversion_rate)
```
