```python
import pandas as pd
```

```python
plays = pd.read_csv('plays.csv')
plays.dtypes
```

```python
plays_4th = plays[plays['down'] == 4]
```

```python
fourth_down = plays_4th[plays_4th['isSTPlay'] == False]
fourth_down
```

```python
fourth_down.loc[fourth_down['yardsToGo'] <= fourth_down['PlayResult'], 'firstDown_Conversion'] = 1 
fourth_down.loc[fourth_down['yardsToGo'] > fourth_down['PlayResult'], 'firstDown_Conversion'] = 0

fourth_down.columns

fourth_down
```

```python
fourth_down.to_csv('plays_4thdowns.csv')
```
