```python
import numpy as np

minutes = np.array(df_2019_fourth['game_seconds_remaining'] // 60).astype(int)
seconds = np.array(df_2019_fourth['game_seconds_remaining'] % 60).astype(int)
GameClock = []
for i in range(len(minutes)):
    GameClock.append(f'{minutes[i]}:{str(seconds[i]).zfill(2)}:00')
```

```python
fourth_down['score_diff'] = fourth_down['HomeScoreBeforePlay'] - fourth_down['VisitorScoreBeforePlay']
```

```python
from datetime import timedelta

def seconder(x):
    mins, secs, millis = map(float, x.split(':'))
    td = timedelta(minutes=mins, seconds=secs, milliseconds=millis)
    return td.total_seconds()
```

```python
fourth_down['game_seconds_remaining'] = fourth_down['GameClock'].apply(seconder)
```
